```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

```

```python
years_SE = [1996, 1997, 1998, 2000, 2002, 2004, 2006, 2011, 2014]
years_EE = [1998, 1999, 2001, 2003, 2006, 2009, 2013, 2018]
version_SE = [0,1,2,3,4,5,6,7,8]
version_EE = [1,2,3,4,5,6,7,8]
years_EEC = [1999, 2001, 2003, 2006, 2009, 2013]
version_EEC = [2,3,4,5,6,7]
classes_SE = [212,504,1520,1824,2991,3279,3793,4024,4240]
classes_EE = [356,427,739,1050,1594,2011]
```

```python
l1=plt.plot(years_SE, version_SE, label='SE Version', marker='*')
l2=plt.plot(years_EE, version_EE, label='EE Version', marker='^')
plt.legend(['SE Version','EE Version'])
plt.ylabel('Version')
plt.xlabel('Year')
plt.show()
```

```python
plt.figure()
l3=plt.plot(years_SE, classes_SE, label='SE Classes', marker='*')
l4=plt.plot(years_EEC, classes_EE, label='EE Classes', marker='^')
plt.legend(['SE Classes','EE Classes'])
plt.ylabel('Classes')
plt.xlabel('Year')
plt.show()
```

```python
plt.figure()
l3=plt.plot(version_SE, classes_SE, label='SE Classes', marker='*')
l4=plt.plot(version_EEC, classes_EE, label='EE Classes', marker='^')
plt.legend(['SE Classes','EE Classes'])
plt.ylabel('Classes')
plt.xlabel('Version')
plt.show()
```
